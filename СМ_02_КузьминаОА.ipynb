Парсинг HTML. XPath

Написать приложение, которое собирает основные новости с сайтов news.mail.ru, lenta.ru, yandex-новости. Для парсинга использовать XPath. Структура данных должна содержать:
[] название источника;
[] наименование новости;
[] ссылку на новость;
[] дата публикации.

```python
from pprint import pprint
from lxml import html
import requests
```

```python
class LentaRuParser:
    host = 'https://lenta.ru'
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) ' +
                             'AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.75 Safari/537.36'}

    def parse(self) -> list:
        news_list = []
        response = requests.get(self.host)
        dom = html.fromstring(response.text)
        news_elements = dom.xpath("//section[contains(@class,'main-page__section')]//a[@class='card-mini _topnews']")
        for element in news_elements:
            news = {
                'title': str(element.xpath("./div[@class='card-mini__text']/h3/text()")[0]),
                'url':  self.host + str(element.xpath("@href")[0]),
                'source': 'LENTA_RU',
                'publication_date': str(element.xpath("./div[@class='card-mini__text']/div[@class='card-mini__info']/time/text()")[0])
            }
            news_list.append(news)
        return news_list
```

```python
class MailRuParser:
    host = 'https://news.mail.ru/'
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) ' +
                             'AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.75 Safari/537.36'}

    def parse(self) -> list:
        news_list = []
        response = requests.get(self.host)
        dom = html.fromstring(response.text)
        news_elements = dom.xpath("//ul[@data-module='TrackBlocks']/li[@class='list__item']")
        for element in news_elements:
            news = {
                'title': str(element.xpath("./span[@class='list__item__inner']/a[@class='list__text']/text()")[0]).replace('\xa0', ' '),
                'url': str(element.xpath("./span[@class='list__item__inner']/a[@class='list__text']/@href")[0]),
                'source': 'MAIL.RU'
            }
            news_list.append(news)
        return news_list
```
